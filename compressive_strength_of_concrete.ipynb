# Checking for outliers

```python
print('Outliers in cement: ', df[((df.cement - df.cement.mean())/df.cement.std()).abs()>3]['cement'].count())
print('Outliers in slag: ', df[((df.slag - df.slag.mean())/df.slag.std()).abs()>3]['slag'].count())
print('Outliers in ash: ', df[((df.fly_ash - df.fly_ash.mean())/df.fly_ash.std()).abs()>3]['fly_ash'].count())
print('Outliers in water: ', df[((df.water - df.water.mean())/df.water.std()).abs()>3]['water'].count())
print('Outliers in superplastic: ', df[((df.superplasticizer - df.superplasticizer.mean())/df.superplasticizer.std()).abs()>3]['superplasticizer'].count())
print('Outliers in coarseagg: ', df[((df.coarse_aggregate - df.coarse_aggregate.mean())/df.coarse_aggregate.std()).abs()>3]['coarse_aggregate'].count())
print('Outliers in fineagg: ', df[((df.fine_aggregate - df.fine_aggregate .mean())/df.fine_aggregate .std()).abs()>3]['fine_aggregate'].count())
print('Outliers in age: ', df[((df.age - df.age.mean())/df.age.std()).abs()>3]['age'].count())
```

# Replacing the outliers by median

```python
for cols in df.columns[:-1]:
  Q1 = df[cols].quantile(0.25)
  Q3 = df[cols].quantile(0.75)
  iqr = Q3 - Q1

  low = Q1-1.5*iqr
  high = Q3+1.5*iqr
  df.loc[(df[cols] < low) | (df[cols] > high), cols] = df[cols].median()
```

**Feature Engineering and Model Building**

```python
#Splitting the data into independent and dependent attributes


#independent and dependent variables
X = df.drop('comp_stregth', axis = 1)
y = df['comp_stregth']
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler() 
Xscaled=sc_x.fit_transform(X)
#Xscaled = X.apply(StandardScaler)
#Xscaled_df = pd.DataFrame(Xscaled, columns=df.columns)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Xscaled, y, test_size= 0.3, random_state= 42)
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression
lr =  LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
#Model Performance on Training Data

lr.score(X_train, y_train)
```

```python
#Model Performance on Test Data

accuracy_lr = lr.score(X_test, y_test)
print(accuracy_lr)
```

```python
from sklearn.metrics import mean_squared_error
mse_lr = mean_squared_error(y_test, y_pred)
print(mse_lr)
```

```python
from sklearn.model_selection import cross_val_score

k = 20

K_results = cross_val_score(lr,X_train, y_train, cv=k,n_jobs=-1)

accuracy=np.mean(abs(K_results)) 
accuracy
```

```python
#Store the accuracy results for each model in a dataframe for final comparison

results_1 = pd.DataFrame({'Algorithm': ['LinearRegression'], 'accuracy': accuracy_lr, 'Kfold': accuracy},index={'1'})
results = results_1[['Algorithm','accuracy','Kfold']]
results
```
